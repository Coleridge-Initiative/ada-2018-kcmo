```python
assignee=pd.read_csv('input/USPTO/assignee.csv')
patent_assignee = pd.read_csv('input/USPTO/patent_assignee.tsv')
kcmo = pd.read_csv('input/KCMO/business_license_2015_2017.csv')
```

```python
kcmo = kcmo.drop_duplicates()
kcmo.describe(include = 'all')
```

```python
patent_assignee
```

```python
assignee = assignee[assignee['organization'].notnull()].copy()
del assignee['type']
del assignee['name_first']
del assignee['name_last']
```

```python
assignee['organization_clean'] = assignee['organization'].str.upper()
kcmo['business_name_clean'] = kcmo['Business Name'].str.upper()

# remove .
assignee['organization_clean'] = assignee['organization_clean'].str.replace('\.', '')
kcmo['business_name_clean'] = kcmo['business_name_clean'].str.replace('\.', '')


# remove .
assignee['organization_clean'] = assignee['organization_clean'].str.replace('\,', '')
kcmo['business_name_clean'] = kcmo['business_name_clean'].str.replace('\,', '')


# replace ' by \s
assignee['organization_clean'] = assignee['organization_clean'].str.replace('\'', ' ')
kcmo['business_name_clean'] = kcmo['business_name_clean'].str.replace('\'', ' ')

# replace - by \s
assignee['organization_clean'] = assignee['organization_clean'].str.replace('-', ' ')
kcmo['business_name_clean'] = kcmo['business_name_clean'].str.replace('-', ' ')

# Replace & by \s&\s
assignee['organization_clean'] = assignee['organization_clean'].str.replace(r'\s?&\s?', ' & ')
kcmo['business_name_clean'] = kcmo['business_name_clean'].str.replace(r'\s?&\s?', ' & ')

# replace Company by CO
assignee['organization_clean'] = assignee['organization_clean'].str.replace('COMPANY', 'CO')
kcmo['business_name_clean'] = kcmo['business_name_clean'].str.replace('COMPANY', 'CO')

# replace National Association by NA
assignee['organization_clean'] = assignee['organization_clean'].str.replace('NATIONAL ASSOCIATION', 'NA')
kcmo['business_name_clean'] = kcmo['business_name_clean'].str.replace('NATIONAL ASSOCIATION', 'NA')

# replace N A by NA
assignee['organization_clean'] = assignee['organization_clean'].str.replace(r'\bN\sA\b', 'NA')
kcmo['business_name_clean'] = kcmo['business_name_clean'].str.replace(r'\bN\sA\b', 'NA')

# replace U S A by USA
assignee['organization_clean'] = assignee['organization_clean'].str.replace(r'\bU\sS\sA\b', 'USA')
kcmo['business_name_clean'] = kcmo['business_name_clean'].str.replace(r'\bU\sS\sA\b', 'USA')

# replace U S by US
assignee['organization_clean'] = assignee['organization_clean'].str.replace(r'\bU\sS\b', 'US')
kcmo['business_name_clean'] = kcmo['business_name_clean'].str.replace(r'\bU\sS\b', 'US')


# When THE is last word, move to front
# TEMP: Remove all THE
assignee['organization_clean'] = assignee['organization_clean'].str.replace(r'^THE\b', '')
assignee['organization_clean'] = assignee['organization_clean'].str.replace(r'\bTHE$', '')
kcmo['business_name_clean'] = kcmo['business_name_clean'].str.replace(r'^THE\b', '')
kcmo['business_name_clean'] = kcmo['business_name_clean'].str.replace(r'\bTHE$', '')

#More radical: remove all special characters:
assignee['organization_clean'] = assignee['organization_clean'].str.replace(r'[^A-Z\s0-9\&]', '')
kcmo['business_name_clean'] = kcmo['business_name_clean'].str.replace(r'[^A-Z\s0-9\&]', '')

#Most radical: remove all non-alpha characters
#assignee['organization_clean'] = assignee['organization_clean'].str.replace(r'[^A-Z]', '')
#kcmo['business_name_clean'] = kcmo['business_name_clean'].str.replace(r'[^A-Z]', '')

# Clean
assignee['organization_clean'] = assignee['organization_clean'].str.strip()
kcmo['business_name_clean'] = kcmo['business_name_clean'].str.strip()
```

```python
kcmo_assignees = pd.merge(kcmo, assignee, how = 'left'
                          , left_on = 'business_name_clean'
                          , right_on = 'organization_clean')
```

```python
kcmo_assignees.describe(include = 'all')
```

```python
kcmo_patents = pd.merge(kcmo_assignees, patent_assignee, how = 'left'
                        , left_on = 'id', right_on = 'assignee_id'
                       )
del(kcmo_patents['assignee_id'])
```

```python
kcmo_patents.describe(include = 'all')
```

```python
kcmo_vars = list(kcmo)
for var in kcmo_vars:
    kcmo_patents[var] = kcmo_patents[var].fillna('')
kcmo_patents['patent_flag'] = kcmo_patents['patent_id'].notnull()
kcmo_patents_grouped = pd.DataFrame({'number_of_patents' : kcmo_patents.groupby(kcmo_vars)
                                     ['patent_flag'].agg('sum')}
                                   ).reset_index()
```

```python
kcmo_patents_grouped.describe(include='all')
```

```python
pd.crosstab(index = kcmo_patents_grouped['number_of_patents'], columns='count')
```

```python
kcmo_patents_grouped[kcmo_patents_grouped['number_of_patents']>25000]
```

```python
pd.crosstab(index = kcmo_patents_grouped['number_of_patents'], columns='count', normalize=True)
```
